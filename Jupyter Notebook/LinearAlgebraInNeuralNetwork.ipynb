# Neural Network

- The design of neural network was based on the human brain.
- Each network has a input neurons(layers), output neurons(layers) and internal neurons(hidden layer).

An input vector $\vec{x} = [x_1 \, x_2 \: x_3 \; \ldots x_n]$ 

A hidden layer vector $\vec{h} = [h_1 \, h_2 \: h_3 \; \ldots h_n]$

An output vector $\vec{y} = [y_1 \, y_2 \: y_3 \; \ldots y_n]$

The lines that connects neurons is denoted by a number called weights.

Since many weights are connected from one layer to the next, so it is denoted by weight matrix.

For weight matrix having element k 
- $w_k$

For ij element of weight matrix k
- $w_{ij}^k$

When working with neural networks we have 2 primary phases: 
- Training and 
- Evaluation

The training phase will include two steps:
- Feedforward
- Backpropagation


## The Feedforward process

### Step-1: Finding hidden layers vector $\vec{h}$



After finding $\vec{h'}$, we need an activation function, which is denoted by greek letter phi: $\phi$

This activation function finalizes the computation of the hidden layer's values.
$\vec{h'} = \phi(\vec{h'})$

or, 

$\vec{h'} = \phi(\vec{x}W^1)$

Here $w_{ij}$ represents the weight matrix values connecting neuron i from the input to neuron j in the hidden layer. 

$h_1 = \phi(x_1.w_{11}+ x_2.w_{21} + \ldots + x_n.w_{n1})$

$h_2 = \phi(x_1.w_{12}+ x_2.w_{22} + \ldots + x_n.w_{n2})$

$h_3 = \phi(x_1.w_{13}+ x_2.w_{23} + \ldots + x_n.w_{n3})$

### Step-2: Finding $\vec{y}$
The final output can be calculated as

$\hat{y} = \phi(\vec{h}W^1)$
